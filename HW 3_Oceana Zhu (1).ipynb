# Section 3: Estimating π Using Monte-Carlo Integration

```python
#Question 11a
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
#Question 11b
def estimate_pi(area_of_circle, area_of_square):
    return 4 * area_of_circle / area_of_square

def absolute_error(estimated_pi):
    return abs(estimated_pi - np.pi)

def inside_unit_circle(x, y):
    return (x**2 + y**2) <= 1
```

```python
#Question 11c
error = []
inside = []
outside = []
number_inside = 0
number_outside = 0
```

```python
#Question 11d

def coord_sampler():
    x = np.random.rand()
    y = np.random.rand()
    return x, y

n_total = 1000

for i in range(n_total):
    x, y = coord_sampler()
    xy_pair = [x, y]
    if (x**2 + y**2) <= 1:
        number_inside += 1
        inside.append(xy_pair)
    else:
        number_outside += 1
        outside.append(xy_pair)
    current_pi_est = estimate_pi(number_inside, number_inside+number_outside)
    
print (current_pi_est)
print (absolute_error(current_pi_est))
    
```

```python
#Question 11e
np_pic = np.array(inside)
np_poc = np.array(outside)

fig1 = plt.figure(figsize=(8,8))
plt.scatter(np_pic[:,0],np_pic[:,1],marker='.',color='r')
plt.scatter(np_poc[:,0],np_poc[:,1],marker='.',color='b')
plt.show()
```

The red points become closer to the origin of the graph, and the blue points become closer to the edges of the graph. This is because as more values of n are used, the better the approximation of the area of the unit circle is.
