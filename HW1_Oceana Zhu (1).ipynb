```python
#a
beef = 11.48
ques = 9.62
salestax = 0.08875

print((beef+ques)*(1+salestax))
```

```python
#b
subtotal = beef+ques
tip = 3.15
print(tip/subtotal)
```

Explanation of process: 
For part a: The sales tax is applied on the bill as a whole before the tax. This means that we need to find the sum of all the goods purchased (beef and ques) and apply a sales tax to it. You need to add 1 to the value of the sales tax before multiplication because you are paying the full value of the beef and ques along with the sales tax. 

For part b: The tip is applied tothe bill before the tax. This means that we need to find the sum of all the goods purchased (subtotal value). We are given the value of the tip, thus, we divide this value by the subtotal to figure out how much of the total the tip was.
